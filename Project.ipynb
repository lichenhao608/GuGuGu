```python
N = len(label)
```

```python
data_dir = 'data/train/'
data = np.zeros((N, 96, 96))
k = 0
for name in label['id'].values:
    name = name + '.tif'
    img_file = data_dir + name
    img = plt.imread(img_file)
    data[k] = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140]) # convert image to grayscale
    
    k += 1
    if k % 1000 == 0:
        print('process {} images'.format(k))
```
